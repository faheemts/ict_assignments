2. Electricity Bill calculator

```python
f = int(input('enter the number of units consumed: '))
t=0
if f <= 100:
  t= f*5
elif f <= 200:
  t= (100*5) + ((f-100)*7)
else:
  t= (100*5) + (100*7) + ((f-200)*10)
print('Your total electricity bill is:',t)
```

4. Palindrome


```python
n=input('enter the number: ')
if n==n[::-1]:
  print('the given number is a palindrome',n)
else:
  print('the given number is not a palindrome',n)
```
