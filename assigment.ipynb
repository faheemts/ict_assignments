Area of a Triangle (Using Heronâ€™s Formula)

```python
e=int(input('enter the a side'))
f=int(input('enter the b side'))
g=int(input('enter the c side'))
s=(e+f+g)/2
h=(s*(s-e)*(s-f)*(s-g))**0.5
print('area of triangle',h)
```
